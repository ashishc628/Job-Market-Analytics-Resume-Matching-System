```python
import pandas as pd
import datetime as dt

today = dt.datetime.now()

def fix_year(date_str):
    if pd.isna(date_str):
        return pd.NaT   # keep null if missing
    
    # Append year
    date_with_year = f"{date_str.strip()} {today.year}"
    
    # Let pandas infer instead of forcing strict format
    date = pd.to_datetime(date_with_year, errors="coerce", dayfirst=True)
    
    # If parsing worked but resulted in future date → assume last year
    if pd.notna(date) and date > today:
        date = date.replace(year=today.year - 1)
    return date

df["Posted_On"] = df["Posted_On"].apply(fix_year)
```

# Natural Laguage Processing

```python
import spacy
print(spacy.__version__)
```

```python
# Load spacy model
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess_text(text):
    """
    Minimal preprocessing: clean, tokenize, remove stopwords, lemmatize
    """
    if pd.isna(text) or text == "":
        return ""
    
    # Basic cleaning
    text = str(text).lower()
    text = re.sub(r'\*+', ' ', text)  # Remove asterisks
    text = re.sub(r'[^\w\s]', ' ', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    
    # Process with spaCy
    doc = nlp(text)
    
    # Tokenize, remove stopwords, lemmatize
    processed_tokens = [token.lemma_ for token in doc 
                       if not token.is_stop and not token.is_punct and len(token.text) > 2]
    
    return ' '.join(processed_tokens)

```

```python
# Apply to your dataframe
df['job_description_processed'] = df['job_description'].apply(preprocess_text)

```

```python
# Define comprehensive skill and domain dictionaries
skills_tech_stack = [
    # Programming Languages
    "python", "java", "javascript", "typescript", "c++", "c#", "php", "ruby", "swift", "kotlin", 
    "go", "rust", "scala", "r", "matlab", "perl", "shell", "bash", "powershell",
    
    # Web Technologies
    "html", "css", "sass", "less", "react", "angular", "vue", "next.js", "nuxt.js", "svelte",
    "jquery", "bootstrap", "tailwind", "material-ui", "node.js", "express", "nestjs", "django",
    "flask", "fastapi", "spring", "laravel", "rails", "asp.net",
    
    # Mobile Development
    "react native", "flutter", "android", "ios", "xamarin", "ionic", "cordova",
    
    # Databases
    "sql", "mysql", "postgresql", "mongodb", "redis", "elasticsearch", "cassandra", "oracle",
    "sqlite", "mariadb", "dynamodb", "cosmosdb", "neo4j", "bigquery", "snowflake", "redshift",
    
    # Data Science & Analytics
    "pandas", "numpy", "scikit-learn", "tensorflow", "keras", "pytorch", "opencv", "nltk",
    "spacy", "matplotlib", "seaborn", "plotly", "jupyter", "anaconda", "spark", "hadoop",
    "kafka", "airflow", "dbt", "looker", "tableau", "powerbi", "qlik", "metabase",
    
    # Cloud & DevOps
    "aws", "azure", "gcp", "docker", "kubernetes", "jenkins", "gitlab", "github", "terraform",
    "ansible", "chef", "puppet", "vagrant", "helm", "istio", "prometheus", "grafana",
    
    # Tools & Platforms
    "git", "jira", "confluence", "slack", "teams", "notion", "trello", "asana", "clickup",
    "figma", "sketch", "adobe", "photoshop", "illustrator", "xd", "invision", "zeplin",
    "postman", "swagger", "insomnia", "charles", "wireshark",
    
    # Testing
    "selenium", "cypress", "jest", "mocha", "pytest", "junit", "testng", "cucumber",
    "appium", "detox", "playwright",
    
    # CRM & Business Tools
    "salesforce", "hubspot", "pipedrive", "zendesk", "freshworks", "servicenow", "workday",
    "sap", "oracle", "microsoft dynamics", "netsuite",
    
    # E-commerce & Marketing
    "shopify", "magento", "woocommerce", "prestashop", "google analytics", "google ads",
    "facebook ads", "linkedin ads", "mailchimp", "sendgrid", "amplitude", "mixpanel"
]

domain_keywords = [
    # Industries
    "fintech", "edtech", "healthtech", "proptech", "regtech", "insurtech", "martech",
    "banking", "finance", "insurance", "healthcare", "education", "real estate", "retail",
    "ecommerce", "e-commerce", "marketplace", "logistics", "supply chain", "manufacturing",
    "automotive", "aerospace", "energy", "utilities", "telecommunications", "media",
    "entertainment", "gaming", "sports", "travel", "hospitality", "food", "agriculture",
    "construction", "mining", "pharmaceutical", "biotechnology", "life sciences",
    
    # Business Models
    "b2b", "b2c", "b2b2c", "saas", "paas", "iaas", "d2c", "marketplace", "platform",
    "subscription", "freemium", "enterprise", "startup", "scaleup", "unicorn",
    
    # Functional Areas
    "product management", "product development", "product strategy", "product marketing",
    "growth", "marketing", "sales", "customer success", "support", "operations",
    "supply chain", "procurement", "hr", "human resources", "recruiting", "talent",
    "finance", "accounting", "legal", "compliance", "risk", "audit", "security",
    "cybersecurity", "data privacy", "gdpr", "hipaa", "pci", "sox",
    
    # Technology Domains
    "artificial intelligence", "machine learning", "deep learning", "computer vision",
    "natural language processing", "nlp", "blockchain", "cryptocurrency", "iot",
    "internet of things", "cloud computing", "edge computing", "microservices",
    "api", "devops", "devsecops", "ci/cd", "automation", "robotics", "ar", "vr",
    "augmented reality", "virtual reality", "mixed reality"
]

soft_skills_keywords = [
    # Communication
    "communication", "verbal communication", "written communication", "presentation",
    "public speaking", "storytelling", "active listening", "cross-cultural communication",
    
    # Leadership & Management
    "leadership", "team leadership", "people management", "project management",
    "program management", "change management", "conflict resolution", "coaching",
    "mentoring", "delegation", "decision making", "strategic thinking",
    
    # Collaboration & Teamwork
    "collaboration", "teamwork", "cross-functional", "stakeholder management",
    "relationship building", "networking", "partnership", "vendor management",
    "client management", "customer facing",
    
    # Problem Solving & Analysis
    "problem solving", "analytical thinking", "critical thinking", "creative thinking",
    "innovation", "design thinking", "systems thinking", "root cause analysis",
    "troubleshooting", "debugging", "research", "data analysis",
    
    # Personal Qualities
    "adaptability", "flexibility", "resilience", "persistence", "self-motivated",
    "proactive", "initiative", "ownership", "accountability", "reliability",
    "attention to detail", "accuracy", "quality focus", "continuous improvement",
    "learning agility", "curiosity", "growth mindset", "emotional intelligence",
    
    # Work Style
    "time management", "prioritization", "organization", "multitasking", "efficiency",
    "productivity", "deadline management", "pressure handling", "fast-paced",
    "startup environment", "agile", "lean", "iterative", "experimental",
    
    # Customer & Business Focus
    "customer centric", "user focused", "business acumen", "commercial awareness",
    "market understanding", "competitive analysis", "revenue focus", "cost conscious",
    "roi", "kpi", "metrics driven", "data driven", "evidence based",
    
    # Interpersonal Skills
    "empathy", "patience", "diplomacy", "negotiation", "persuasion", "influence",
    "consensus building", "facilitation", "mediation", "cultural sensitivity"
]

def extract_skills(text):
    """Extract technical skills from text"""
    doc = nlp(text.lower())
    found = []
    for skill in skills_tech_stack:
        if skill.lower() in doc.text:
            found.append(skill)
    return list(set(found))

def extract_domains(text):
    """Extract domain/industry keywords from text"""
    doc = nlp(text.lower())
    found = []
    for domain in domain_keywords:
        if domain.lower() in doc.text:
            found.append(domain)
    return list(set(found))

def extract_soft_skills(text):
    """Extract soft skills from text"""
    doc = nlp(text.lower())
    found = []
    for skill in soft_skills_keywords:
        if skill.lower() in doc.text:
            found.append(skill)
    return list(set(found))

def extract_entities(text):
    """Extract named entities using spaCy NER"""
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

# Apply extraction functions to the dataframe
# Note: Using 'product_description' column as specified
df['extracted_skills'] = df['job_description_processed'].apply(extract_skills)
df['extracted_domains'] = df['job_description_processed'].apply(extract_domains)
df['extracted_soft_skills'] = df['job_description_processed'].apply(extract_soft_skills)
df['entities'] = df['job_description_processed'].apply(extract_entities)

# Combine all extractions for comprehensive analysis
def combine_all_extractions(row):
    """Combine all extracted information"""
    return {
        'technical_skills': row['extracted_skills'],
        'domains': row['extracted_domains'],
        'soft_skills': row['extracted_soft_skills'],
        'named_entities': row['entities']
    }

df['comprehensive_extraction'] = df.apply(combine_all_extractions, axis=1)

# Save cleaned data
df.to_csv("cleaned_jobs_comprehensive_ner.csv", index=False)

print("✅ Data cleaned, preprocessed with comprehensive NER, and saved to cleaned_jobs_comprehensive_ner.csv")
print("\n📊 Sample results:")
for i in range(min(3, len(df))):
    print(f"\n--- Job {i+1} ---")
    if 'title' in df.columns:
        print(f"Title: {df.iloc[i]['title']}")
    if 'company' in df.columns:
        print(f"Company: {df.iloc[i]['company']}")
    print(f"Technical Skills: {df.iloc[i]['extracted_skills'][:5]}...")  # Show first 5
    print(f"Domains: {df.iloc[i]['extracted_domains'][:3]}...")  # Show first 3
    print(f"Soft Skills: {df.iloc[i]['extracted_soft_skills'][:3]}...")  # Show first 3

```

```python
import pandas as pd
import ast

# Load your dataset
df_flatten = df

# Function to safely convert to list
def to_list(x):
    if isinstance(x, list):       # Already a list
        return x
    elif isinstance(x, str):
        x = x.strip()
        if x == "" or x == "[]":
            return []
        try:
            return ast.literal_eval(x)  # Convert string to list
        except:
            # If literal_eval fails, treat as a single-item list
            return [x]
    else:
        return []

# Apply safely to all list-columns
for col in ["skills", "domains", "soft_skills"]:
    df_flatten[col] = df_flatten[col].apply(to_list)

# Now explode
skills_expanded = df_flatten.explode("skills")
domains_expanded = df_flatten.explode("domains")
soft_skills_expanded = df_flatten.explode("soft_skills")

# Optional: Save CSVs for Tableau
skills_expanded.to_csv("jobs_skills_flat.csv", index=False)
domains_expanded.to_csv("jobs_domains_flat.csv", index=False)
soft_skills_expanded.to_csv("jobs_soft_skills_flat.csv", index=False)
```

```python
import pandas as pd
from ast import literal_eval



# -----------------------
# 2. Create Jobs Table
# -----------------------
jobs_df = df[[
    "Title", "Company", "Location", "Type", "Level", "Posted_On", "Job_URL", "job_description_processed"
]].reset_index(drop=True)

jobs_df["job_id"] = jobs_df.index + 1  # assign job IDs
jobs_df.to_csv("jobs.csv", index=False)

print("✅ jobs.csv created:", jobs_df.shape)


```

```python
# -----------------------
# 3. Create Skills Table + Mapping
# -----------------------
skills_exploded = df[["skills"]].explode("skills").dropna().reset_index()
skills_exploded = skills_exploded.rename(columns={"index": "job_id"})  # job_id will match row index

# Unique skills
unique_skills = pd.DataFrame(skills_exploded["skills"].unique(), columns=["skill_name"]).dropna()
unique_skills["skill_id"] = range(1, len(unique_skills) + 1)

# Mapping job ↔ skill
job_skills = skills_exploded.merge(unique_skills, left_on="skills", right_on="skill_name")[["job_id", "skill_id"]]
job_skills["job_id"] = job_skills["job_id"] + 1  # shift to match job_id from jobs.csv

# Save
unique_skills.to_csv("skills.csv", index=False)
job_skills.to_csv("job_skills.csv", index=False)

print("✅ skills.csv & job_skills.csv created:", unique_skills.shape, job_skills.shape)

```

```python
# -----------------------
# 4. Create Domains Table + Mapping
# -----------------------
domains_exploded = df[["domains"]].explode("domains").dropna().reset_index()
domains_exploded = domains_exploded.rename(columns={"index": "job_id"})

unique_domains = pd.DataFrame(domains_exploded["domains"].unique(), columns=["domain_name"]).dropna()
unique_domains["domain_id"] = range(1, len(unique_domains) + 1)

job_domains = domains_exploded.merge(unique_domains, left_on="domains", right_on="domain_name")[["job_id", "domain_id"]]
job_domains["job_id"] = job_domains["job_id"] + 1

# Save
unique_domains.to_csv("domains.csv", index=False)
job_domains.to_csv("job_domains.csv", index=False)

print("✅ domains.csv & job_domains.csv created:", unique_domains.shape, job_domains.shape)

```

```python
# -----------------------
# 5. Create Soft Skills Table + Mapping
# -----------------------
soft_exploded = df[["soft_skills"]].explode("soft_skills").dropna().reset_index()
soft_exploded = soft_exploded.rename(columns={"index": "job_id"})

unique_soft = pd.DataFrame(soft_exploded["soft_skills"].unique(), columns=["soft_skill_name"]).dropna()
unique_soft["soft_skill_id"] = range(1, len(unique_soft) + 1)

job_soft = soft_exploded.merge(unique_soft, left_on="soft_skills", right_on="soft_skill_name")[["job_id", "soft_skill_id"]]
job_soft["job_id"] = job_soft["job_id"] + 1

# Save
unique_soft.to_csv("soft_skills.csv", index=False)
job_soft.to_csv("job_soft_skills.csv", index=False)

print("✅ soft_skills.csv & job_soft_skills.csv created:", unique_soft.shape, job_soft.shape)
```
